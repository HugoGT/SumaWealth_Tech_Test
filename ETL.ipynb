## Email ETL

```python
df = pd.read_csv('base_data/email.csv')
df
```

```python
df.info()
```

```python
df['Customer_Location'].fillna('.', inplace=True)
df['Customer_Location'] = df['Customer_Location'].apply(lambda x: x.replace('.', random.choice(['A', 'B', 'C', 'D', 'E', 'F', 'G'])))
```

```python
df['Customer_Location'].value_counts()
```

```python
df['Total_Past_Communications'].fillna(0, inplace=True)
df['Total_Links'].fillna(0, inplace=True)
df['Total_Images'].fillna(0, inplace=True)
```

```python
df['Total_Past_Communications'] = df['Total_Past_Communications'].astype(int)
df['Total_Links'] = df['Total_Links'].astype(int)
df['Total_Images'] = df['Total_Images'].astype(int)
```

```python
df['Subject_Hotness_Score'] = df['Subject_Hotness_Score'].apply(lambda x: round(x))
```

```python
new_names = {column: column.lower() for column in df.columns}
df.rename(columns=new_names, inplace=True)
```

```python
df.rename(columns={'email_campaign_type': 'campaign_id'}, inplace=True)
df['campaign_id'] = df['campaign_id'].apply(lambda x: (x * 0) + random.randint(1, 22))
df['campaign_id'].value_counts()
```

```python
mapping = {1: 'soft', 2: 'aggressive'}
df['email_type'] = df['email_type'].replace(mapping)
```

```python
df.rename(columns={
    'subject_hotness_score': 'customer_score',
    'total_past_communications': 'past_communications'
}, inplace=True)
```

```python
df.drop(columns=['time_email_sent_category', 'email_source_type', 'customer_score', 'customer_location'], inplace=True)
```

## Users ETL

```python
df = pd.read_csv('base_data/tiktok_profiles.csv')
df.head(5)
```

```python
df.info()
```

```python
df['create_time'] = df['timestamp']
df['account_id'] = df['id']
```

```python
df = df[['account_id', 'nickname','awg_engagement_rate', 'comment_engagement_rate', 'like_engagement_rate', 'likes', 'create_time']]
```

```python
df
```

```python
df['nickname'] = df['nickname'].apply(lambda x: x.replace(x, '_'))
```

```python
df = pd.concat([df] * 4, ignore_index=True)
```

```python
df
```

```python
df['account_id'] = df['account_id'].apply(lambda x: random.randint(10000, 40000))
df.drop_duplicates(subset=['account_id'], inplace=True)
df
```

```python
def random_date(day):
    # Define el rango de fechas
    start_date = datetime(2022, 2, 12)
    end_date = datetime(2023, 6, 30)

    # Calcula la diferencia en días entre las fechas de inicio y fin
    days_difference = (end_date - start_date).days

    # Genera una fecha aleatoria dentro del rango
    random_date = start_date + timedelta(days=random.randint(0, days_difference))

    return random_date.date()
```

```python
df['create_time'] = df['create_time'].apply(random_date)
df['create_time']
```

```python
import requests
import time

names = []

url = 'https://randomuser.me/api/'

for _ in range(3740):
    response = requests.get(url)

    data = response.json()['results'][0]['name']

    name = data['first'] + ' ' + data['last']

    names.append(name)

    time.sleep(0.5)

names
```

```python
df['nickname'] = names
df.rename(columns={'nickname': 'full_name', 'account_id': 'customer_id'}, inplace=True)
```

```python
max(df['awg_engagement_rate'])
```

```python
df.info()
```

```python
aggregated_data['past_communications'] = aggregated_data['past_communications'].apply(lambda x: round(x))
aggregated_data
```

```python
df = df.merge(aggregated_data, on='customer_id', how='left')
df
```

```python
def engagement(n):
    if n > 10:
        return n / random.randint(10, 20)
    elif n > 1:
        return n + random.uniform(1, 2)
    else:
        return n / random.uniform(0.8, 0.1)
```

```python
df['awg_engagement_rate'] = df['awg_engagement_rate'].apply(engagement)
df['like_engagement_rate'] = df['like_engagement_rate'].apply(engagement)
df['comment_engagement_rate'] = df['comment_engagement_rate'].apply(engagement)
df
```
